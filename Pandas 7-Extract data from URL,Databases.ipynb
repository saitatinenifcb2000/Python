# Merging Dataframes

```python
df1=pd.DataFrame({'key':['b','c','a','a','b','c','a'],'data1':range(7)})
```

```python
df1
```

```python
df2=pd.DataFrame({'key':['a','b','d'],'data2':range(3)})
```

```python
df2
```

```python
pd.merge(df1,df2)
```

#only gives o/p for common values only in our case its b & a
#You need to have 1 common column to merge dataframe on that column.

# Stackplot

```python
days=[1,2,3,4,5]
sleeping=[7,8,6,4,5]
eating=[1,2,4,3,2]
working=[10,12,14,15,11]
playing=[1,4,5,2,3]
plt.stackplot(days,sleeping,eating,working,playing,colors=['r','b','k','y'],labels=['sleep','eat','work','play'])
plt.legend(loc='best')
plt.xlabel('Days')
plt.ylabel('hours')
plt.show()
```

# Subplot

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
```

```python
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.subplot(221)
plt.plot(t1,f(t1),'bo',t2,f(t2),'r')
plt.subplot(222)
plt.plot(t2,np.cos(2*np.pi*t2),'y')
plt.xlabel('Graph 2')
plt.show()
```

```python
t3=np.arange(1,5.0,0.5)
t4=np.arange(1,5.0,0.02)
plt.subplot(221)
plt.plot(t3,'b',t4,'r')
plt.show()
```

#t3 & t4 are arange values which displays arnage in increasing order
#for ex t3 has only 6 or 7 values on x axis as its (1-5) with interval of 0.5.
#t4 has 200+ intervals on x-axis as its rannge is (1-5) with inteval of 0.02

```python
#subplot 221 stands for 2 for plotting 2 graphs,2 for horizontal & 1 for first graph.
```

```python
t3=np.arange(1,5,1)
t4=np.arange(1,5,1)
plt.subplot(221)
plt.plot(t3,t4,'.')
plt.subplot(222)
plt.plot(t4,'.')
plt.xlabel('X-axis')
plt.show()
```

#Its difficult to have categorical data in ML. sO WE CHANGE THESE CATEGORIES DATA INTO SOme kind of numerical data

```python
sex=pd.get_dummies(titanic_data['Sex'],drop_first=True)
sex.head(4)
```

#get dummies converts the categorical value into 0 &1
#here for sex it converts the column sex into 0 &1
#drop_first drops the first column.

```python
pcl=pd.get_dummies(titanic_data['Pclass'],drop_first=True)
pcl.head(4)
```

```python
embark=pd.get_dummies(titanic_data['Embarked'],drop_first=True)
embark.head(4)
```

```python
#concatinating above to original data
```

```python
td1=pd.concat([titanic_data,sex,pcl,embark],axis=1)
```

#removing categorical data for ML algo

```python
td1.drop(columns=['PassengerId','Pclass','Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```
